```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(2, 3, tight_layout=True)

y00 = [1, 2, -5, 2]
ax[0,0].plot(y00)

x01 = np.linspace(0, 10, 100)
y01 = 2 + 2 * np.sin(2 * x01)
ax[0, 1].plot(x01, y01, linewidth=2.0)
ax[0, 1].set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))

x02 = np.random .normal(15, 5, 2000)
ax[0, 2].hist(x02)

np.random.seed(3)
x10 = 4 + np.random.normal(0, 2, 60)
y10 = 4 + np.random.normal(0, 2, len(x10))
ax[1, 0].scatter(x10, y10)

plt.show()
```
